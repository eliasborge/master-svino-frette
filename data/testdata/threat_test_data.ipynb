```python


with open("VideoCommentsThreatCorpus.txt", 'r', encoding='utf-8') as f:
    lines = f.readlines()
    print(lines)
```

```python
current_index = 0
previous_block_end = 0

comment_blocks = []
for line in lines:
    if(line == "\n"):
        test = 0
        comment_blocks.append(lines[previous_block_end:current_index])
        previous_block_end = current_index
        current_index += 1
    else:
        current_index +=1
  
print(comment_blocks[0:3])
```

```python

list_of_comments = []
for comment_block in comment_blocks:
    processed_comment = {}
    for comment in comment_block:
        if(comment == "\n"):
            comment_block.remove(comment)
    comment_details = comment_block[0]
    comment_content = comment_block[1:]
    processed_comment["details"] = comment_details
    processed_comment["content"] = comment_content
    list_of_comments.append(processed_comment)
    
print(list_of_comments[0:3])
```

```python
for comment in list_of_comments:
   

    parts = comment['details'].split(", ")
    video_num = parts[0].split("#")[1] 
    comment_num = parts[1].split("#")[1]  
    comment_id = video_num + "_" + comment_num

    comment['id'] = comment_id
    
    comment_content = comment['content']
    flags = []
    joined_content = ""
    for line in comment_content:
        flags.append(line[0])
        line_without_flags = line[2:]
        joined_content += line_without_flags
    comment['flags'] = flags
    comment['content'] = joined_content

print(list_of_comments[0:3])
```

```python
import pandas as pd

df = pd.DataFrame(columns=['id', 'flags', 'content'])

for comment in list_of_comments:
    row = pd.DataFrame([{
        'id': comment['id'],
        'flags': comment['flags'],
        'content': comment['content']
    }])
    df = pd.concat([df, row], ignore_index=True)

print(df.head())

```

```python
print(df.head())
```

```python
def sum_flags(flags):
    return 1 if any(flag == '1' for flag in flags) else 0

df['flags'] = df['flags'].apply(sum_flags)
print(df.head())
```

```python
df['video_num'] = df['id'].str.split('_').str[0]
df['num_posts'] = df.groupby('video_num')['video_num'].transform('count')
```

```python
print(df.head())
```

```python
# Ensure the 'flags' column contains lists of flags
df['flags'] = df['flags'].apply(lambda x: [x] if isinstance(x, int) else x)

grouped_df = df.groupby('video_num').agg({
    'content': lambda x: [comment for comment in x],  # List of comments
    'flags': lambda x: [flag for sublist in x for flag in sublist],
    'id': lambda x: ', '.join(x)  # Include IDs
}).reset_index()

grouped_df['content_length'] = grouped_df['content'].apply(len)

print(grouped_df.iloc[13])
```

```python
import ast

MAX_CHARS = 30000  # Define limit for splitting

def splitTopic(topic: pd.DataFrame) -> pd.DataFrame:
    if topic.empty:
        raise ValueError("Input DataFrame is empty.")
    
    # Extract video number (same for all rows in the topic)
    original_video_num = topic['video_num']

    content_list = topic['content']
    id_list = topic['id'].split(',')
    flags_list = topic['flags']

    num_posts = len(content_list)
    total_content_length = sum(len(c) for c in content_list)

    if num_posts != len(id_list) or num_posts != len(flags_list):
        raise ValueError(
            f"Mismatch: num_posts={num_posts}, len(id_list)={len(id_list)}, len(flags_list)={len(flags_list)}"
        )

    # If total content fits within MAX_CHARS, return as a single topic without splitting
    if total_content_length <= MAX_CHARS:
        return topic

    # Compute optimal splits
    optimal_splits = max(1, -(-total_content_length // MAX_CHARS))  # Ceiling division
    split_size = num_posts // optimal_splits
    remainder = num_posts % optimal_splits

    split_topics = []
    start_idx = 0

    for i in range(optimal_splits):
        end_idx = start_idx + split_size + (1 if i < remainder else 0)
        sub_ids = id_list[start_idx:end_idx]
        sub_content = content_list[start_idx:end_idx]
        sub_flags = flags_list[start_idx:end_idx]

        split_topics.append({
            'video_num': f"{original_video_num}_{i+1}",  # Append split number
            'id': ','.join(sub_ids),
            'content': sub_content,
            'flags': sub_flags,
            'content_length': sum(len(c) for c in sub_content),
            'num_posts': len(sub_content)
        })

        start_idx = end_idx

    return pd.DataFrame(split_topics)


```

```python
new_grouped_df = pd.DataFrame(columns=grouped_df.columns)
for i, topic in grouped_df.iterrows():
    new_topic = splitTopic(topic)
    print(new_topic["video_num"])
    new_grouped_df = pd.concat([new_grouped_df, new_topic], ignore_index=True)

```
