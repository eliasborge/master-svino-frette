# Kennedy et al. Gab Hate Corpus breakdown

Labels: 
* HD = Violating Human Dignity
* VO = Vulgarity and/or Offensive Language
* CV = **Calls for violence**

Only half a percent of the posts are labeled with calls for violence, but can possibly be used as a testset, or some form of benchmark test.

```python
import numpy as np
import pandas as pd
import os

file_path = '.././data/kennedy/ghc_train.tsv'
df = pd.read_csv(file_path, sep='\t')
```

```python
cols = df.columns
print("Number of rows: ", len(df))
print("Columns:")
for col in cols:
    print("    "+col) 
print("\n")

print("Number of missing values in each column:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each label
label_counts = df[['hd', 'cv', 'vo']].sum()

# Count the rows where all labels are zero
unlabeled_count = (df[['hd', 'cv', 'vo']].sum(axis=1) == 0).sum()
label_counts['unlabeled'] = unlabeled_count

# Plot the distributions
label_counts.plot(kind='bar')
plt.xlabel('Labels')
plt.ylabel('Count')
plt.title('Distribution of Labels')
plt.show()
```

```python
cv_messages = df[df['cv'] == 1]['text']
print(len(cv_messages), " - percentage: ", len(cv_messages)/len(df))
```
