# Gaikwad Twitter ISIS/White supremacy Dataset


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('.././data/gaikwad/MIWS/MIWS.csv')
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Row count
row_count = df.shape[0]
print(f"Row count: {row_count}")

# Column count
column_count = df.shape[1]
print(f"Column count: {column_count}")

# Word count in 'Labels' column
df['word_count'] = df['Labels'].apply(lambda x: len(str(x).split()))
total_word_count = df['word_count'].sum()
print(f"Total word count in 'Labels' column: {total_word_count}")

# Average word length in 'Labels' column
df['avg_word_length'] = df['Labels'].apply(lambda x: np.mean([len(word) for word in str(x).split()]))
avg_word_length = df['avg_word_length'].mean()
print(f"Average word length in 'Labels' column: {avg_word_length}")

# All possible values in 'Labels' column
unique_labels = df['Labels'].unique()
print(f"All possible values in 'Labels' column: {unique_labels}")

# Missing values
missing_values = df.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")

# Plotting average word length
plt.figure(figsize=(10, 6))
sns.histplot(df['avg_word_length'], bins=30, kde=True)
plt.title('Distribution of Average Word Length in Labels')
plt.xlabel('Average Word Length')
plt.ylabel('Frequency')
plt.show()

# Plotting count of each label
plt.figure(figsize=(12, 8))
sns.countplot(y='Labels', data=df, order=df['Labels'].value_counts().index)
plt.title('Count of Each Label')
plt.xlabel('Count')
plt.ylabel('Labels')
plt.show()
```
