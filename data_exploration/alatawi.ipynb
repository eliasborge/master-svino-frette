# Alatawi Twitter white supremacy data

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('.././data/alatawi/WhiteSupremacistDataset.csv')
```

```python
# Row count
row_count = len(df)

# Average post length
average_post_length = df['input.text'].apply(len).mean()

# Missing values
missing_values = df.isnull().sum()

# Columns
columns = df.columns

# Display the results
print(f"Row count: {row_count}")
print(f"Average post length: {average_post_length}")
print("Missing values:")
print(missing_values)
print("Columns:")
print(columns)
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of 'Voting and Final Labels'
label_distribution = df['Voting and Final Labels'].value_counts()
label_distribution.plot(kind='bar')

# Add title and labels
plt.title('Distribution of Voting and Final Labels')
plt.xlabel('Labels')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Calculate word lengths for each post
df['word_length'] = df['input.text'].apply(lambda x: len(x.split()))

# Plot the distribution of word lengths
plt.figure(figsize=(10, 6))
plt.hist(df['word_length'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Post Word Lengths')
plt.xlabel('Word Length')
plt.ylabel('Frequency')
plt.show()
```
